### BigQuery Query문 구조


```sql
WITH CTE AS (
    SELECT
        ...
    FROM Table
)

SELECT
    a.col3,
    b.col4,
    COUNT(DISTINCT a.id) AS cnt
FROM table_a AS a
LEFT JOIN table_b AS b
ON a.id = b.id
WHERE
    a.col3 >= 3
GROUP BY
    a.col3
    b.col4
HAVING
    cnt >= 2
ORDER BY cnt DESC
LIMIT 10
```

### 현황 파악
- 회사 또는 제품의 현재 상황을 파악
    - 전체 상황을 파악하고, 앞으로 무엇을 하면 좋을지 생각하기 위한 기반 지식

    $\rightarrow$ 이 과정에서 가설 도출 가능
- 현황을 분석한 후, 다양한 이해 관계자의 이야기를 듣고, 해결할 수 있는 Action Item을 함께 만들고 성과를 확인

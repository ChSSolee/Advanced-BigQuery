## 퍼널 분석 (Funnel)

<br>

### 퍼널 분석의 큰 흐름

<img src='img/02-16.png' width=600>

<br>

### 퍼널의 정의
- **하나의 퍼널에서 다음 퍼널로 얼마나 전환되는가를 파악**
- 퍼널은 목적에 맞게 구성 가능
    - **특정 페이지를 하나의 퍼널로 볼 수 있고, 특정 페이지의 묶음을 퍼널로 정의 가능**
- 대표적인 퍼널 : 회원 가입 퍼널, 온보딩 퍼널, 결제 퍼널 등
    - 또는 회원 가입 내 에서도 - 정보 입력 퍼널, 인증 퍼널 등

<br>

### 퍼널을 정의하는 방법

<br>

#### 이벤트가 명시적으로 존재하는 경우
- 예) 메인 화면 진입, 구매 완료
- 이벤트의 조합으로 사용

<br>

#### 명시적으로 존재하진 않으나, 모든 페이지의 `view` 이벤트가 존재하는 경우
- 페이지의 `view`를 기준으로 퍼널을 정의
- **`view` 이벤트, 커스텀 이벤트를 모두 활용**

<br>

### 퍼널의 종류

<br>

#### Open 퍼널
- 유저의 이벤트 행동 순서에 상관없이 집계
- 모든 단계에서 시작될 수 있다고 가정
    - 예) 원 클릭 결제가 있다면, 특정 순서를 넘길 수 있음

<img src='img/02-17.png' width=600>

<br>

#### Closed 퍼널 (대부분의 분석)
- 유저가 정해진 단계를 순서대로 거쳐야 집계
- 엄격한 순서를 거쳐야 함
- **다음 단계가 항상 이전 단계보다 작거나 같음**

<img src='img/02-18.png' width=600>

<br>

#### Closed 퍼널과 Open 퍼널 비교
- **Closed 퍼널은 유저의 로그 데이터를 기반으로, 앞선 퍼널의 이벤트가 먼저 진행되었는지 확인**
- **Open 퍼널은 해당 이벤트가 존재하는지를 파악**

    **$\rightarrow$ 어떤 퍼널을 사용할 것인가?**

    **$\rightarrow$ 제품에 대한 이해가 필수**

<br>

- **제품을 잘 이해해야 하는 이유** = 어떤 퍼널을 사용할 것인가?가 결정
    - 제품에 따라 다르게 결정
    - 예) 온보딩 과정, 결제 과정에서 Skip 하는 기능이 존재할 수 있음
- **어떤 앱은 자유롭게 이동할 수 있으나, 어떤 앱은 순서가 엄격하게 고정되어 있는 경우도 존재**
    - 웹은 자유롭게 이동 가능 (URL Direct 접근)
    - 앱도 Deep Link로 직접 접근하는 경우가 존재

<br>

#### 퍼널 선택의 고민
- **Open 퍼널**
    - 유연한 유저의 여정 : 다양하게 진입할 수 있는 경우
    - 광범위한 분석 : 시작 위치와 관계없이 모든 사용자의 상호 작용을 파악하고 싶은 경우
    - 최초에 큰 그림을 보고 싶은 경우

<br>

- **Closed 퍼널**
    - 유저의 흐름이 고정된 경우
    - 모든 단계가 순서대로 완료되어야 하는 경우
    - 순차적으로 해야하며, 앞선 조건이 엄격한 경우

<br>

### 퍼널의 개선

<br>

#### 어떤 퍼널을 먼저 개선해야 하는가?

<img src='img/02-19.jpg' width=400>

- 상황, 전략적인 방향에 따라 다름
- **전환율이 낮은곳부터 하는 것이 많이 추천되나**
    
    고객이 우리의 제품을 확실히 좋아하는 상황을 먼저만들고, 습관을 형성한 후에 전환율을 올리는 것이 좋을수 도 있음

- **제품의 근본적인 부분을 손보지 않고 특정 전환율만 올리려고 하면 근본적인 해결이 아님**
    
    $\rightarrow$ **­실무관점에서 결국 모두다 해야함**
    
    $\rightarrow$ 둘다 같이 할수 있는 부분이 있는지, ROI가 괜찮은 곳에 먼저 투자하는 것도 방법

<br>

### 지표 집계 단위

<br>

#### 대표적인 방법
1. 유저(User) 별로 집계할 것인가?
2. 페이지뷰(PageView) 별로 집계할 것인가?
3. 세션(Session) 별로 집계할 것인가?

+ **여러차원(Dimension)으로 집계**
    - 시간축 : 일자별/주차별/월별등
    - 지역정보
    - 사용자인구통계

<br>

### 지표 집계 단위 예시

| user_id | 날짜 | 시간 | 활동 |
| - | - | - | - |
| A | 20XX-XX-01 | 9:00 | 앱 실행 |
| A | 20XX-XX-01 | 9:01 | 추천 목록 조회 |
| A | 20XX-XX-01 | 9:03 | 영상 시청 (1분) |
| B | 20XX-XX-01 | 10:00 | 앱 실행 |
| B | 20XX-XX-01 | 10:02 | 검색 |
| B | 20XX-XX-01 | 10:05 | 영상 시청 (5분) |

<br>

- **유저별 집계** : 고유한 **사용자 수**를 `COUNT(DISTINCT user_id)`

    $\rightarrow$ 사용자 : 2명

<br>

- **페이지뷰 별 집계** : 페이지뷰 이벤트가 발생한 **횟수**를 `COUNT`
    - 앱에서는 `page_view` 대신 `screen_view`

    $\rightarrow$ screen_view 총 6회

<br>

- **세션 별 집계**
    - 세션 : 웹이나 앱에 사용자가 접속한 시점부터 이탈할 때의 활동  
    - 동일한 유저가 세션 종료후, 다른 시간에 접속하면 하나의 새로운 세션이 시작
    - 일정시간 (일반적으로 30분)동안 유저의 활동이 없으면 세션이 종료된 것으로 간주
    - 유저가 서비스를 명시적으로 종료 (예:로그아웃)하면 세션이 종료로 간주

- **세션을 어떻게 만드는가?**
    - **session_id가 존재하는 경우**
        
        $\rightarrow$ 해당 세션을 사용할 수 있음. 단, session이 어떤 기준으로 만들어지는지 확인 필요 (30분동안 활동이 없다면 이탈로 간주 등)
    - **session_id 가없는경우**
        
        $\rightarrow$ 사용자의 로그 기반으로 세션을 직접 생성
        
        - 세션을 만들때 윈도우 함수를 사용해서 미활동의 시간을 임의로 설정할 수 있음

- **일자가 바뀌는 시점에 세션을 어떻게 처리하는가?**
    - 일자가 넘어가는 시점에 걸쳐서 세션이 존재하는 경우, 처리가 어려울 수 있음
    - 예) 20XX년 X월 5일 23시58분에 접속해서, X월 6일 1시에 세션이 종료된 경우엔 X월 5일에 포함해야하는가?

    $\rightarrow$ 기준에 따라 다르고, 정의하기 나름

    - 아예 12시에 모든 세션을 종료시키는 것도 방법이고, 혹은 사용자가 갑자기 떨어지는 시점 기준 (예:새벽 3시)을 토대로 하는 것도 방법

- **세션을 꼭 사용해야 하는가?**
    - 어떤 기준을 사용하는지는 해결하고자하는 문제의 목적에 따라 다름
    - 하루에 접속이 많은 서비스라면 세션을 확인하는 것이 좋음
    - 하루에 반복해서 쓰는 비율이 적다면, 세션의 집계를 굳이 하지 않고, 일자별-유저별 집계를 하는 것이 더 좋을 수도 있음
    - 세션을 직접 만들면 쿼리가 복잡해지기 때문에, 작업하는 시간이 더 소요될 수도 있다는 점을 인지
    - 판단이 어렵다면 일자 단위, 세션 단위 등을 구한후에 어떻게 해석이되는가 살펴보고 결정 
    
        (데이터를 추출하고보면 계속 비슷하게 가는 경우도 있고, 아예 다른 해석이 되는 경우도 존재)

- **“이탈”이라는이벤트가명시적으로발생할까?**
    - 앱, 웹 서비스에선 이탈 이벤트가 명확하지 않은 경우가 많음
    - 로그아웃등은 명확 하지만, 사용할때마다 로그아웃을 하지않는 경우도존재 $\rightarrow$ 따라서 유저의 행동이 없는경우를 이탈로 정의하는 경우가 많음
    - 앱 활성화 여부, 사용자 인터랙션등을 종합적으로 고려하여 세션기준을 만들어야함

### 퍼널 쿼리 작성
- 오픈 퍼널
- `user_pseudo_id` 집계 기준

<br>

#### 퍼널 정의 
- **6개의 이벤트**
    - **Welcome $\rightarrow$ Home $\rightarrow$ Food Category $\rightarrow$ Restaurant $\rightarrow$ Cart $\rightarrow$ Click Payment**

<br>

<img src='img/02-20.jpg' width=600>

<br>

<img src='img/02-21.jpg' width=600>

<br>

#### 각 퍼널의 유저수를 집계

```sql
WITH base AS (
  SELECT
    event_date,
    event_timestamp,
    event_name,
    user_id,
    user_pseudo_id,
    platform,

    MAX(IF(ep.key = "firebase_screen", ep.value.string_value, NULL)) AS firebase_screen,
    MAX(IF(ep.key = "session_id", ep.value.string_value, NULL)) AS session_id
  FROM advanced.app_logs
  CROSS JOIN UNNEST(event_params) AS ep
  WHERE
    event_date BETWEEN "2022-08-01" AND "2022-08-18"
  GROUP BY ALL
), filter_event_and_concat_event_and_screen AS (
  SELECT
    * EXCEPT(event_name, firebase_screen, event_timestamp),
    CONCAT(event_name, "-", firebase_screen) AS event_name_with_screen,
    DATETIME(TIMESTAMP_MICROS(event_timestamp), 'Asia/Seoul') AS event_datetime
  FROM base
  WHERE
    event_name IN ("screen_view", "click_payment")
)

SELECT
  event_date,
  event_name_with_screen,
  CASE 
    WHEN event_name_with_screen = "screen_view-welcome" THEN 1
    WHEN event_name_with_screen = "screen_view-home" THEN 2
    WHEN event_name_with_screen = "screen_view-food_category" THEN 3
    WHEN event_name_with_screen = "screen_view-restaurant" THEN 4
    WHEN event_name_with_screen = "screen_view-cart" THEN 5
    WHEN event_name_with_screen = "click_payment-cart" THEN 6
  ELSE NULL
  END AS step_number,
  COUNT(DISTINCT user_pseudo_id) AS cnt
FROM filter_event_and_concat_event_and_screen
GROUP BY ALL
HAVING step_number IS NOT NULL
```

## 퍼널 분석 의사 결정

<br>

#### 현황 파악
- 서비스에 대한 전반적인 큰 그림 파악
- 현재 주요 이벤트는 얼마나 발생하고 있는가?
- 퍼널 별로 얼마나 사용자가 있는가?
    - 퍼널을 어떻게 정의할 것인가?
    - 일자별로 달라지는가? (시간의 흐름)
    - 특히 더 빠지는 구간 있는가? 등

<br>

#### 퍼널 데이터 해석
- 단순히 그래프만 보는 것이 아닌 숫자를 보면서 계속 질문

    - 정말 이런가? 맞다면 앋른 데이터가 어떻게 나와야 할까?
    - 전체로 봤을 때와 일자별로 나눌 떄의 결과가 같을까?
    - 다른 경로는 어떤가?
    - 일자별 또는 공휴일의 패턴이 존재하는가?

<br>

#### 퍼널 관련 아이디어
- 퍼널 단계를 줄이기 (Gap 줄이기)
- 퍼널 순서를 바꾸기:
    - 회원 가입 전에 체험하고, 체험의 어느 단계에서 회원 가입 또는 로그인하도록 설계
    - 결제 직전에 추가 상품을 추천해서 업셀링
    - 장바구니에 담기 전에 할인 프로모션 안내, 할인 가격 적용해서 노출

<br>

- **너무 넓게 퍼널을 정의하면, 그 사이에 발생하는 것 들에 대한 것을 알 지 못함**
    - 큰 퍼널로 감을 잡고 $\rightarrow$ 더 쪼개서 페이지 별 퍼널을 파악하는 것도 좋은 관점

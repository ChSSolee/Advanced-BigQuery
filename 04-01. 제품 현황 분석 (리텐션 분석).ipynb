### 리텐션 (Retention)
- 시간이 흐르면서 닷 제품 (앱, 웹)을 사용하는지를 측정하는 지표
- 시간이 흐르면서, 계쏙 제품을 다시 사용하는 리텐션이 높은 서비스도 있고,

    1회만 사용해서 재사용이 적은 리텐션이 낮은 서비스도 존재

    - **고객의 특성 (유료, 무료) 마다 리텐션 그래프가 다르게 나올 수 있음**

- **PMF (Product Market Fit)을 찾았는가?를 생각할 때, 리텐션 커브를 그려보고 판단**
    - PMF : 제품이 고객의 핵심 문제 상황을 효과적으로 해결할 경우
    - **평평한 곳이 유지가되면 PMF를 찾았다고 간주**

<br>

#### 리텐션 분석의 고민 포인트
1. **Retain의 정의가 필요**
2. **접속 주기가 서비스마다 상이할 수 있음**
    - 예) 금융앱의 리텐션 vs 부동산 앱의 리텐션
        
        $\rightarrow$ 금융앱의 리텐션보다, 부동산 앱의 리텐션 주기가 길 수 밖에 없음

        $\rightarrow$ 부동산 앱의 리텐션을 늘리기 위해 부동산 커뮤니티 기능 추가 사례 존재

    - **하나의 제품에서 여러 페르소나가 있는 경우, 다른 주기로 리텐션을 볼 수도 있음**

3. **생각한 주기가 실제랑 같은지 파악**
4. **Retenton은 결과 지표 (Output Metric) $\rightarrow$ 원인을 파악하기 위해서는 구체적으로 쪼개보는 작업이 필요**

<br>

#### 제품의 핵심
- 우리 제품은 고객의 어떤 문제를 해결하려고 하는가?
- 핵심 이벤트는 무엇인가?
    - 제품이 고도화되어있으면, 하나의 제품에도 여러 핵심 이벤트가 존재할 수 있음
    - 예) 
        - 소셜 앱 : 피드를 본다
        - 음식 주문 앱 : 음식을 주문
        - 택시 앱 : 택시를 탑승
        - 이커머스 앱 : 원하는 제품을 주문
    
    - **처음 이벤트를 구할 때, 앱 접속을 시작으로, 점점 구체적인 이벤트 (핵심 이벤트)의 리텐션을 파악해야 함**

        $\rightarrow$ **앱 접속은 목표가 아닌 과정**

        - **예) 앱 접속이 코어 이벤트인 경우? $\rightarrow$ 광고 플랫폼 $\rightarrow$ 접속하는 것이 곧 노출과 직결 $\rightarrow$ DAU의 중요성과 직결**
   

 
<br>

#### 제품의 사용주기
1. **예상하는 사용 주기를 떠올리기**
    - 고객은 이 문제를 얼마나 겪는가? 어떤 빈도로 겪는가?

2. **예상한 주기와 실제 주기가 일치하는지 파악**
    - 예) 최근 1주/1달의 접속일 확인
        - 빠른 의사결정이 필요? $\rightarrow$ 1주
    - 예) 제품 사용 주기에 대한 분포 확인

<img src='img/04-02.png' width=400>


3. **평균 사용 주기 확인**
    - 예) 2번 이상 사용한 사람들의 평균 사용 주기
    - 만약 유저의 첫 사용 주기의 간격을 알고 싶다면 첫 사용과 다음 사용 사이의 간격을 계산

### 리텐션의 종류

<br>

#### Classic 리텐션 (N day 리텐션)
- **사용자가 최초로 '참여'한 Day 0 이후, 각 Day N 별로 한 번 더 '참여' 했는지 계산**
- 제일 보수적인 접근이며, 낮은 값이 나올 수 있음
- **사용주기가 짧거나 정기적으로 사용하는 경우 Daily로 많이 활용**
    - **예) 마케팅 캠페인을 한 후, 반응을 하루 단위로 보고 싶은 경우**

<br>

#### Range 리텐션 (Bracket 리텐션)
- **Day N이 아닌 특정 Range 내에서 더 '참여'했는지 계산**
- **제일 유연하고, 제일 실용적인 관점**
    - **예) 주차별 범위를 만들어서 계속 사용하는지 확인하고 싶은 경우**

<br>

#### Rolling 리텐션 (Unbounded 리텐션)
- **Day N 이후, 한 번이라도 '참여'한 경우를 그 동안 모두 참여했다고 간주**
- 가장 낙관적인 접근이며, 높은 값이 나올 수 있음
- **사용주기가 긴 제품의 경우에 활용**
    - **예) 언제 다시 돌아올 것 같은지 예상 되지 않는 경우**

<br>

- **Classic 리텐션**
    - **Day 1 리텐션 (2024-01-02)** : 2/3

    - Day 0 ~ 1 : 두 날짜 중 1번이라도 Retaion

        $\rightarrow$ **Day 7 ~ 8 리텐션 : 2/3**
- **Rolling 리텐션**
    - **Day 1 Retention (2024-01-02)** : 3/3 
        
        $\rightarrow$ **2024-01-02에 C는 Retaion이 없지만, 이후의 기간 동안 1번이라도 Retaion이 존재하기 때문** 
    - **Day 7 Retention** : 2/3
    - **Day 8 Retention** : 0/3

### N Day Retention
- **Core Event : 앱 접속**
- **Classic 리텐션 (Daily)**

```sql
WITH base AS (
  SELECT
    DISTINCT 
      user_id,
      event_name,
      DATETIME(TIMESTAMP_MICROS(event_timestamp), 'Asia/Seoul') AS event_datetime,
      DATE(DATETIME(TIMESTAMP_MICROS(event_timestamp), 'Asia/Seoul')) AS event_date,
      user_pseudo_id
  FROM advanced.app_logs
  WHERE
    event_date BETWEEN "2022-08-01" AND "2022-11-03"
)

, first_data AS (
  SELECT
    user_pseudo_id, 
    MIN(event_date) AS first_date
  FROM base
  GROUP BY user_pseudo_id
)

, event_data AS (
  SELECT
    user_pseudo_id,
    event_date
  FROM base
  GROUP BY ALL
)

, retain_base AS (
  SELECT
    fd.user_pseudo_id,
    fd.first_date,
    ed.event_date,
    DATE_DIFF(ed.event_date, fd.first_date, DAY) AS diff_of_day
  FROM first_data AS fd
  LEFT JOIN event_data AS ed
  ON fd.user_pseudo_id = ed.user_pseudo_id
)

, user_counts AS (
  SELECT
    diff_of_day,
    COUNT(DISTINCT user_pseudo_id) AS user_cnt
  FROM retain_base
  GROUP BY diff_of_day
)

-- , first_day_user_count AS (
--   SELECT
--     user_cnt AS first_day_user_cnt
--   FROM user_counts
--   WHERE
--     diff_of_day = 0
-- )

-- SELECT
--   uc.diff_of_day,
--   uc.user_cnt,
--   fduc.first_day_user_cnt,
--   ROUND(SAFE_DIVIDE(uc.user_cnt, fduc.first_day_user_cnt), 3) AS retention_rate
-- FROM user_counts AS uc
-- CROSS JOIN first_day_user_count AS fduc
-- ORDER BY diff_of_day

-- 혹은 윈도우 함수를 사용한 방법
SELECT
  *,
  ROUND(SAFE_DIVIDE(user_cnt, first_day_user_cnt), 3) AS retention_rate
FROM (
  SELECT
    diff_of_day,
    user_cnt,
    FIRST_VALUE(user_cnt) OVER(ORDER BY diff_of_day ASC) AS first_day_user_cnt
  FROM user_counts
)
```

### Weekly Retention
- **각 날짜를 주차별로 (혹은 월요일의 날짜로 처리)** 
  - `WEEK` 혹은 `DATE_TRUNC`

<br>

- `WEEK` vs `ISO_WEEK`
  - 주 번호를 계산할 때 사용할 수 있는 함수
  - `WEEK` : **일요일이 주의 첫 날로 간주, 1월 1일이 속한 주가 1주차**
  - `ISO_WEEK` : **월요일이 주의 첫 날로 간주. ISO 8601 국제 표준에 따라 정의. 목요일이 속한 주를 기준으로 주 번호를 지정**
    - 목요일이 속하면 그 주의 4일 이상이 포함되기 때문
    - 연도의 첫 목요일이 있는 주부터 1주차
- 2022년 1월 1일(토요일) 
  
  - `WEEK` : 2022년 1주차
  - `ISO_WEEK` 2021년 52주차

<br>

```sql
WITH base AS (
  SELECT
    DISTINCT
      user_id,
      user_pseudo_id,
      event_name,
      DATE(DATETIME(TIMESTAMP_MICROS(event_timestamp), 'Asia/Seoul')) AS event_date,
      DATETIME(TIMESTAMP_MICROS(event_timestamp), 'Asia/Seoul') AS event_datetime
  FROM advanced.app_logs
  WHERE
    event_date BETWEEN "2022-08-01" AND "2022-11-03"
)

, first_week_and_diff AS (
  SELECT
    *,
    DATE_DIFF(event_week, first_week, WEEK) AS diff_of_week
  FROM (
    SELECT
      DISTINCT 
        user_pseudo_id,
        DATE_TRUNC(MIN(event_date) OVER(PARTITION BY user_pseudo_id), WEEK(MONDAY)) AS first_week,
        DATE_TRUNC(event_date, WEEK(MONDAY)) AS event_week
    FROM base
  )
)

, user_counts AS (
  SELECT
    diff_of_week,
    COUNT(DISTINCT user_pseudo_id) AS user_cnt
  FROM first_week_and_diff
  GROUP BY diff_of_week
)


SELECT
  *,
  ROUND(SAFE_DIVIDE(user_cnt, first_week_user_cnt), 2) AS retention_rate
FROM (
  SELECT
    diff_of_week,
    user_cnt,
    FIRST_VALUE(user_cnt) OVER(ORDER BY diff_of_week ASC) AS first_week_user_cnt
  FROM user_counts
)
```
